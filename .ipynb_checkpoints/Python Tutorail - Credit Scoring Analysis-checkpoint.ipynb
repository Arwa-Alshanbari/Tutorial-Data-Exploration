Generating a visualization to display the proportion of missing values in `credit_scoring`

1. `plt.figure(figsize=(10,6))`: This line is using Matplotlib to set up a new figure with a specified size.

2. `sns.displot()`: is a figure-level function for drawing distribution plots.

3. `credit_scoring.isna()`: This is checking the `credit_scoring` DataFrame for missing values.

4.  The `value_name="missing"` argument specifies the name of the new column in the melted DataFrame that will contain the values which are the `True`/`False` indications of missing data from `credit_scoring.isna()`.

5. `y="variable"`: In the melted DataFrame, there is a column named `variable` that contains the original column names from `credit_scoring`.

6. `hue="missing"`: This parameter means that the plot will use different colors (hues) to distinguish between `True` (missing values) and `False` (non-missing values).

7. `multiple="fill"`: This tells Seaborn to stack the bars for `True` and `False` values of the `missing` column on top of each other and normalize them so that the sum is 1 (or 100%). This will effectively show the proportion of missing values for each variable.

##### Visualizing data

Visualizing data is useful because it allows you to see relationships in data in a fast, intuitive way. It's beneficial in exploring data and deciding what to dig into next because it can point to places where there may be significant patterns. In pandas, the `.plot()` method allows you to create several different types of charts with the DataFrame and Series objects.

````python
# Visualizing Missing Data
plt.figure(figsize=(10,6))
sns.displot(
 credit_scoring.isna().melt(value_name="missing"),
    y="variable",
    hue="missing",
    multiple="fill",
    aspect=1.25)
````

From the analysis of missing values, it's observed that approximately 10% of the data related to the customers' employment duration `days_employed` and their monthly income `total_income` is missing. We must choose whether to drop the rows with missing data, considering the percentage is relatively low, or to fill in the gaps with the mean or median values

 **sns.countplot** is designed to create a bar chart that displays the frequency or count of observations in each category of a categorical variable. It's a quick way to visualize the distribution of categorical data and is especially useful for comparing the number of occurrences or frequency across different categories.

````python
# understanding Data - Data Exploration
# Count the occurrences and sort
order = credit_scoring['family_status'].value_counts().index
# Plotting with seaborn
sns.countplot(x='family_status', data=credit_scoring, order=order)
plt.title('Family Status Counts')
plt.ylabel('Count')
plt.xlabel('Family Status')
````

This chart is showing the counts of individuals by their family status. Each bar represents a category of family status such as "married", "civil partnership", "unmarried", "divorced", and "widow/widower".

The height of each bar corresponds to the number of occurrences in the dataset for that particular family status. For example, the "married" category has the tallest bar, indicating that it is the most common family status among the individuals in this dataset, followed by "civil partnership" and the others in decreasing order.

For example, there are more married women than married men in this list.

-----------------------------------------------------------------

**grouped.unstack()** : when you have data grouped by two categories (like color and size), grouped.unstack() rearranges the data so you can see all the subcategories of one category side by side for each main category. It turns a list-like view into a table-like view.

````python
# understanding Data - Data Exploration
grouped = credit_scoring.groupby(['family_status_id', 'gender']).size()
unstacked = grouped.unstack('gender').fillna(0)
ax = unstacked.plot(kind='bar', figsize=(10, 6))
ax.set_title('Distribution of Family Status by Gender')
ax.set_xlabel('Family Status ID')
ax.set_ylabel('Count')
ax.legend(title='Gender')
plt.xticks(rotation=45)
plt.show()
````

The visualization is a bar chart that shows the number of people by their marital status and gender. Each group of bars represents a different marital status, coded by numbers 0 to 4. Here's what each number represents:

0: Married
1: Civil partnership
2: Widow/widower
3: Divorced
4: Unmarried

Each marital status has up to three bars in different colors, showing the count for females (F), males (M), and a non-binary or unspecified gender (XNA). The height of each bar shows how many people there are for each gender within each marital status category.

For instance, married females (0, F) have the highest bar, indicating they are the largest group in the dataset, while there's a very small number represented by XNA in the married category, and so on for the other statuses.

**`Histograms`** are particularly useful when dealing with large datasets, as they can provide a visual summary of the data. They can help us understand the underlying frequency distribution of a set of continuous or discrete data. This can be particularly useful when dealing with data like age groups, where understanding the distribution can provide valuable insights.
Creating a histogram in pandas is straightforward thanks to the `hist()` function.

This function provides a quick way to visualize the distribution of data in a pandas DataFrame or Series.

The `hist()` function in pandas uses a default number of bins, which is 10. However, you can customize the number of bins according to your needs. The bins parameter in the `hist()` function is used to specify the number of bins you want in your histogram.
Here we want to increase the number of bins to 100 to see more details.

<br>
<br>
<br>

Next, you will see a histogram where the x-axis represents the total monthly income spread across different ranges (bins), and the y-axis represents the number of people (frequency) whose income falls within each range. This visualization is useful for understanding the distribution of income within the dataset, such as determining if the income data is skewed, has outliers, or follows a particular pattern.

````python
# Change the size of font and figure
plt.rcParams['font.size'] = 12
plt.rcParams["figure.figsize"] = [8, 4]

#  total monthly income spread across different bins
credit_scoring['total_income'].hist(bins=100)


# Setting the plot labels and title
plt.title ('monthly income')
plt.xlabel('month')
plt.ylabel('income $')
````
